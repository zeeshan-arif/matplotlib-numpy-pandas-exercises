```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("fivethirtyeight")
```

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

```python
# basic pie chart, having two portions

slices = [120, 80]

plt.pie(slices)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [120, 80]
labels = ["Sixty", "Forty"]

plt.pie(slices, labels=labels, wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [120, 80, 30, 20]
labels = ["Sixty", "Forty", "Extra1", "Extra2"]
colors = ["blue", "red", "yellow", "green"]

plt.pie(slices, labels=labels, colors=colors, wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# here pie chart looks more clean. Pie chart is for 5

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

plt.pie(slices, labels=labels, colors=colors, wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# here pie chart looks more clean. Pie chart is for 5

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# here pie chart looks more clean. 3-D type chart

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True, wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# rotate the chart by 90 degrees

# here pie chart looks more clean. Pie chart is for 5

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90, wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# rotate the chart by 90 degrees

# here pie chart looks more clean. Pie chart is for 5

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90, autopct="%1.1f%%", wedgeprops={"edgecolor": "black"})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
