```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use("seaborn")
```

```python
data = pd.read_csv("data3.csv")
ages = data["Age"]
dev_salaries = data["All_Devs"]
py_salaries = data["Python"]
js_salaries = data["JavaScript"]
```

```python
plt.plot(ages, py_salaries, label="Python")
plt.plot(ages, js_salaries, label="JavaScript")

plt.plot(ages, dev_salaries, color="#444444", linestyle='--', label="All Devs")

plt.legend()

plt.title("Median Salary (USD)by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

Everytime we are doing something with our plot, we are using "plt" object.

fig, ax => fig is a container. Wereas ax is single plot

A figure can have multiple plots

#### lets create the same plot that we have above, instead of using pyplot object we are using fig, ax

```python
# with fig, ax our title, xlabel, ylabel changes to set_title, set_xlabel, set_ylabel

fig, ax = plt.subplots()

ax.plot(ages, py_salaries, label="Python")
ax.plot(ages, js_salaries, label="JavaScript")

ax.plot(ages, dev_salaries, color="#444444", linestyle='--', label="All Devs")

ax.legend()

ax.set_title("Median Salary (USD)by Age")
ax.set_xlabel("Ages")
ax.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

#### We have the same plot as above....task accomplished

```python
# we have one figure with two different plots

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)

ax1.plot(ages, dev_salaries, color="#444444", linestyle='--', label="All Devs")

ax2.plot(ages, py_salaries, label="Python")
ax2.plot(ages, js_salaries, label="JavaScript")

ax1.legend()

ax1.set_title("Median Salary (USD)by Age")
ax1.set_ylabel("Median Salary (USD)")

ax2.legend()

ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

```python
# we have one figure with two different plots
# sharing x-axis

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(ages, dev_salaries, color="#444444", linestyle='--', label="All Devs")

ax2.plot(ages, py_salaries, label="Python")
ax2.plot(ages, js_salaries, label="JavaScript")

ax1.legend()

ax1.set_title("Median Salary (USD)by Age")
ax1.set_ylabel("Median Salary (USD)")

ax2.legend()

ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

```python
# we have one figure with two different plots
# sharing y-axis

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharey=True)

ax1.plot(ages, dev_salaries, color="#444444", linestyle='--', label="All Devs")

ax2.plot(ages, py_salaries, label="Python")
ax2.plot(ages, js_salaries, label="JavaScript")

ax1.legend()

ax1.set_title("Median Salary (USD)by Age")
ax1.set_ylabel("Median Salary (USD)")

ax2.legend()

ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

```python
# we now have two different figures

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, dev_salaries, color="#444444", linestyle='--', label="All Devs")

ax2.plot(ages, py_salaries, label="Python")
ax2.plot(ages, js_salaries, label="JavaScript")

ax1.legend()

ax1.set_title("Median Salary (USD)by Age")
ax1.set_ylabel("Median Salary (USD)")

ax2.legend()

ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```
